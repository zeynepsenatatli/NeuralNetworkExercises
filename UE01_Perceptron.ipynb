## Logic Gates as Perceptron

```python
import numpy as np

inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
```

### AND

```python
weight = np.array([1,1])
bias = -1.5
```

```python
def and_perceptron(x, weight, bias):

    z = np.dot(weight, x) + bias

    if z > 0:
        return 1
    else:
        return 0
```

```python
for i in inputs:
    print(str(i) + " -> " + str(and_perceptron(i, weight, bias)))
```

### OR

```python
weight = np.array([1,1])
bias = -0.5
```

```python
def or_perceptron(x, weight, bias):

    z = np.dot(weight, x) + bias

    if z > 0:
        return 1
    else:
        return 0
```

```python
for i in inputs:
    print(str(i) + " -> " + str(or_perceptron(i, weight, bias)))
```

### NOT

```python
weight = np.array([-1])
bias = 1
```

```python
def not_perceptron(x, weight, bias):

    z = np.dot(weight, x) + bias

    if z > 0:
        return 1
    else:
        return 0
```

```python
for i in [0,1]:
    print(str(i) + " -> " + str(not_perceptron(i, weight, bias)))
```
